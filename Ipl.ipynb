```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('matches.csv')
```

```python
df.head()
```

```python
#Looking at the number of rows and columns in the dataset
df.shape
```

```python
#Getting the frequency of most man of the match awards
df['player_of_match'].value_counts()
```

```python
#Getting the top 10 players with most man of the match awards
df['player_of_match'].value_counts()[0:10]
```

```python
#Getting the top 5 players with most man of the match awards
df['player_of_match'].value_counts()[0:5]
```

```python
list(df['player_of_match'].value_counts()[0:5].keys())
```

```python
#Making barplot for the top 5 players with most man of the match awards
plt.figure(figsize=(8,5))
plt.bar(list(df['player_of_match'].value_counts()[0:5].keys()), list(df['player_of_match'].value_counts()[0:5]), color='g')
plt.show()
```

```python
#Getting the frequency of result column
df['result'].value_counts()
```

```python
#Finding out the number of toss wins w.r.t each team
df['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won battling first
batting_first = df[df['win_by_runs']!=0]
```

```python
#Looking at the head
batting_first.head()
```

```python
#Making a histogram
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of runs")
plt.xlabel("Runs")
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
#Making a barplot for top 3 teams with most wins after batting first(in barplot 1st parameter is categorical then 2nd is numerical)
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()), list(batting_first['winner'].value_counts()[0:3]), color=['blue','yellow','orange'])
plt.show()
```

```python
#Making a pie chart(In pie chart 1st parameter is numerical and 2nd paramater is categorical)
plt.figure(figsize=(8,8))
plt.pie(list(batting_first['winner'].value_counts()), labels=list(batting_first['winner'].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```

```python
#Exracting those records where a team has won after batting second
batting_second = df[df['win_by_wickets']!=0]
```

```python
#Looking at the head
batting_second.head()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize=(5,5))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
#Finding out the frequency of number of wins w.r.t each time after batting second
batting_second['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after batting second
plt.figure(figsize=(5,5))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()), list(batting_second['winner'].value_counts()[0:3]), color=['purple','blue','red'])
plt.show()
```

```python
#Making a pie chart for distribution of most wins after batting second
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()), labels=batting_second['winner'].value_counts().keys(), autopct="%0.1f%%")
plt.show()
```
